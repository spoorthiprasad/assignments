```python
from scipy import stats
import numpy as np
import pandas as pd
```

```python
durability = [1.13, 1.55, 1.43, 0.92, 1.25, 1.36, 1.32, 0.85, 1.07, 1.48, 1.20, 1.33, 1.18, 1.22, 1.29]
durability
```

```python
len(durability)
```

```python
sample_mean =np.mean(durability)
sample_mean
```

```python
sample_std=np.std(durability)
print(sample_std)
```

### **confidence interval using t distribution**

```python
durability_ci=stats.t.interval(0.99,df=14,loc=sample_mean,scale=sample_std)
print(durability_ci)
print("99% of confidence interval is : ",np.round(durability_ci,4))
```

## **confidence interval using normal distribution**

```python
durability_ci=stats.norm.interval(0.99,loc=sample_mean,scale=sample_std)
print(durability_ci)
print("99% of confidence interval is : ",np.round(durability_ci,4))
```
