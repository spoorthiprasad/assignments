```python
import mlxtend
from mlxtend.frequent_patterns import apriori,association_rules
from mlxtend.preprocessing import TransactionEncoder
```

```python
df=pd.read_csv("/content/Online retail.csv")
df
```

```python
df["Transaction"]=df.iloc[:,0].apply(lambda x:x.split(","))
df
```

```python
te = TransactionEncoder()
te_ary = te.fit_transform(df['Transaction'])
df_encoded = pd.DataFrame(te_ary, columns=te.columns_)
```

### Apriori Algorithm

```python
freq_items=apriori(df_encoded,min_support=0.05,use_colnames=True)
freq_items
```

```python
rules=association_rules(freq_items,metric="lift",min_threshold=0.1)
rules
```

```python
rules.sort_values("lift",ascending=False)
```

```python
rules[(rules['lift'] > 1) & (rules['confidence'] > 0.3)]
```
